2️⃣ What does “learns all the unique categories” mean?

When you call:

ohe.fit_transform(df)


The fit part does this:

It looks at your column (say gender) and finds all possible categories:

gender → Male, Female

cough → Yes, No

city → Jaipur, Delhi, Mumbai

It remembers these categories. This is important because when you later give new data to the model, it knows how to encode it properly.

Without learning the unique categories first, it wouldn’t know how many new columns to create or what 0/1 means.

Example:

Original column:

Gender :
Male
Female
Male

fit → learns unique categories: [Male, Female]

transform → converts into 0/1 columns:

Gender_Female :
0
1
0

```python
x_train_sc=sc.fit_transform(x_train)# fit = lrearn the parameter  and transform means apply on that data
```

```python
x_train_new = pd.DataFrame(x_train_sc,columns=x_train.columns)
```
