```python
import pandas as pd
import numpy as np
import sklearn
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
```

```python
df=pd.read_csv("/content/Social_Network_Ads - Social_Network_Ads.csv")
```

```python
df.sample(5)
```

```python
df.isnull().sum()
```

```python
df=df.drop(columns="User ID")
```

```python
x=df.drop(columns="Purchased")
```

```python
y=df["Purchased"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.compose import ColumnTransformer
```

```python
transformer=ColumnTransformer([
    ("tnf1",StandardScaler(),["Age","EstimatedSalary"]),
    ("tnf2",OneHotEncoder(sparse_output=False,drop="first"),["Gender"])
],remainder="passthrough")
```

```python
transformer.fit(x_train)
```

```python
x_train_new=transformer.transform(x_train)
```

```python
x_test_new=transformer.transform(x_test)
```

```python
x_train_new=pd.DataFrame(x_train_new)
x_test_new=pd.DataFrame(x_test_new)

```
