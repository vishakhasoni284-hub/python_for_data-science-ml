```python
#Countplot: Counts frequency of each unique value, numeric or categorical. it only takes one column. just like value counts
#Barplot: Summarizes a numeric variable for each category. It takes two columns
#countplot doesn’t care if your column is numeric or categorical.
#What matters is: it treats the values as categories, not numbers.
#So even if the column is numeric, countplot will count how many times each unique value appears.
#Bar plots are used to compare categorical data using a numerical value.
#Bar plot = categorical (x) + numerical (y)
```

```python
sns.set(style='whitegrid')
plt.figure(figsize=(9,5))
sns.countplot(data=df,x='Rating')#Think of it like df['Rating'].value_counts(), but it directly plots it as a bar chart.
plt.title('Distribution of ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

```python
from textblob import TextBlob
```

```python
def textblob_sentiments_analysis(review):
  sentiment=TextBlob(review).sentiment # sentiment is inbuilt in textblob , (.sentiment ) gives a Sentiment object that means a object becomes sentimrnts so we wrote sentiment as a variable name
  if sentiment.polarity >0.1 :
    return 'positive'
  elif sentiment.polarity < -0.1:
    return 'negative'
  else:
    return 'neutral'
```

```python
df['Sentiment']=df['Review'].apply(textblob_sentiments_analysis)
```

```python
df.sample(5)
```

```python
sentiment_distribution =df ['Sentiment'].value_counts()
sentiment_distribution
```

```python
plt.figure(figsize=(9,5))
sns.barplot(x=sentiment_distribution.index,
            y=sentiment_distribution.values)
plt.title('Distribution of sentiments')
plt.xlabel("Sentiment")
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df,
              x='Rating',
              hue='Sentiment')
plt.xlabel("Rating")
plt.ylabel('Count')
plt.legend(title='Sentiment')
plt.show()
```
