```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("/content/titanic - titanic.csv")
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.columns
```

```python
df['Survived'].value_counts()
```

```python
from matplotlib import color_sequences
sns.countplot(x=df['Survived'],color='red')# countplot tells how many in each category
```

```python
#sns.countplot() is a Seaborn plot that automatically counts the frequency of
#each category of a column and displays it as a bar chart. It is specially designed for
#categorical data. On the other hand, df['col'].value_counts().plot(kind='bar')
#first needs value_counts() to calculate the frequencies and then draws a simple
#bar graph using Pandas. Countplot has better styling and directly works with
#raw data, whereas the Pandas bar plot works on already-summarised data.
```

```python
#CountPlot = Automatic Counting
#Bar Plot = Manual Counting
```

```python
df['Survived'].value_counts().plot(kind='bar')# it is of pandas seaborn also has its own bar plot this ons is of pandas
```

```python
df['Survived'].value_counts().plot(kind='pie',autopct='%.2f')
```
