**Linear SVM classification**

````python
import pandas as pd
import numpy as np
from sklearn import datasets
from sklearn.pipeline import Pipeline
from sklearn.svm import LinearSVC
from sklearn.preprocessing import StandardScaler
````

````python
iris=datasets.load_iris()
iris
````

````python
# it has four features
# 0 --> sepel length
# 1 --> sepal width
# 2 --> petal length
# 3 --> petal width

# we will only take petal ones that is 2 and 3
````

````python
x = iris["data"][ :, (2,3)] #[][] = first access the dict, then slice the array.
y = (iris["target"]==2).astype(np.float64)
#int64 stores whole numbers only
#float64 stores decimals too

#Since ML math (gradients, losses, probabilities) almost always produces decimal
#values, starting with float64 avoids conversion issues down the line.


````

````python
svm_clf=Pipeline([
    ("std",StandardScaler()),
    ("linear_svc",LinearSVC(C=1,loss="hinge"))
])
````

````python
svm_clf.fit(x,y)
````

````python
svm_clf.predict([[5.5, 1.7]])
````

SVM doesn't ask *"how likely is spam?"*
It only asks *"which side of the boundary is this on?"*


**Nonlinear SVM Classification**


1. Pipeline containing a
PolynomialFeatures in linear svc


````python
from sklearn.datasets import make_moons
from sklearn.preprocessing import PolynomialFeatures
import matplotlib.pyplot as plt

````

````python
x, y = make_moons(n_samples=100, noise=0.15, random_state=42)
````

n_samples = 100

# Step 1 — splits 100 points into 2 groups
upper_moon = 50 points   # label → y = 0
lower_moon = 50 points   # label → y = 1

# Step 2 — places each group along a circle arc mathematically
upper_moon coordinates:
   point1 → (0.0,  0.5)   # calculated from circle formula
   point2 → (0.3,  0.9)
   point3 → (-0.3, 0.8)
   ...50 points

lower_moon coordinates:
   point1 → (1.0, -0.5)   # calculated from circle formula
   point2 → (0.7, -0.3)
   ...50 points

# Step 3 — adds noise (small random shifts)
(0.0, 0.5) + noise → (0.02, 0.48)  # slightly moved
```

---

### So Who Does What?
```
n_samples=100  →  creates 100 points total
                  splits into 50 + 50

math inside    →  calculates exact coordinates
                  using circle arc formula
                  assigns y=0 or y=1 automatically

noise=0.15     →  slightly shifts each coordinate
                  so points aren't perfectly on arc

random_state=42 → fixes which random shifts are applied
                  so you get same result every time

````python
poly_svm_clf=Pipeline([
    ("poly_features",PolynomialFeatures(degree=3)),
    ("scalar",StandardScaler()),
    ("model",LinearSVC(C=10,loss='hinge'))
])
````

````python
poly_svm_clf.fit(x,y)
````

2. Polynomial Kernel


````python
from sklearn.svm import SVC
poly_kernel_svm_clf = Pipeline([
("scaler", StandardScaler()),
("svm_clf", SVC(kernel="poly", degree=3, coef0=1, C=5))
])
poly_kernel_svm_clf.fit(x, y)
````

3. Using RBF Kernal

````python
rbf_kernel_svm_clf = Pipeline([
("scaler", StandardScaler()),
("svm_clf", SVC(kernel="rbf", gamma=5, C=0.001))
])
rbf_kernel_svm_clf.fit(x, y)
````
